```python
# Age - Income - Total_Spending - NumWebPurchases - NumStorePurchases - NumWebVisitsMonth - Recency
features = ["Age", "Income", "Total_Spending", "NumWebPurchases", "NumStorePurchases", "NumWebVisitsMonth", "Recency"]
X = df[features].copy()
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(2, 10), wcss, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Within-cluster sum of squares (WCSS)")
plt.show()
```

```python
kmeans = KMeans(n_clusters=6)
df["Cluster"] = kmeans.fit_predict(X_scaled)
df["Cluster"]
```

```python
df.head()
```

```python
cluster_summary = df.groupby("Cluster")[features].mean()
cluster_summary
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)   
pca_data = pca.fit_transform(X_scaled)
df["PCA1"], df["PCA2"] = pca_data[:,0], pca_data[:,1]
```

```python
# silhouette score
from sklearn.metrics import silhouette_score
score = silhouette_score(X_scaled, df["Cluster"])
score
```

High-income customers who spend a lot and shop actively.

Older, stable customers with moderate income and balanced spending.

Cluster 2:

Young digital shoppers who prefer online purchases and visit the website frequently.

Cluster 3:

Inactive customers who haven’t purchased recently and show very low engagement.

Cluster 4:

Low-income customers with low spending and occasional buying behavior.

Cluster 5:

High-income customers who spend less but have potential to be converted into premium buyers.
